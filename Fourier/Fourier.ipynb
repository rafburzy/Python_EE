## Fourier transform
#### visualization using polar plots

This notebook is inspired by the 3Blue1Brown YouTube movie on visual introduction to Fourier transform available 
<a href="https://youtu.be/spUNpyF58BY">here</a>.

```python
import numpy as np
import ipywidgets as widgets
from ipywidgets import interact
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('ggplot')
```

```python
t = np.linspace(0, 0.4, 1000)
f = 50
a = 28*np.sin(2*np.pi*f*t)

plt.figure(figsize=(15,6))
plt.plot(t, a)
plt.xlim(0, 0.2);
```

```python
def plotPolar(f):
    g = a * np.exp(-1j*2*np.pi*f*t)
    plt.figure(figsize=(10,10))
    plt.polar(np.angle(g), np.abs(g));
```

```python
plotPolar(17)
```

```python
interact(plotPolar, f = widgets.IntSlider(value=7, min=1, max=70))
```
